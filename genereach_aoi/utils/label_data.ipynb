```python
def labeling(path, target_dir, goal):
  files = sorted([os.path.join(path + target_dir + goal, x) for x in os.listdir(path+target_dir+goal) if x.endswith(".jpg")])
  length = len(files)
  print(length)
  id = 0
  zeros = 0
  ones = 0
  last_result = 0
  # input("please:")
  while id < length:
    pic = Image.open(files[id])
    plt.imshow(pic)
    plt.show()
    
    plt.gcf().canvas.draw()
    
    t_or_f = input(f"please input your judgment {id+1}: \n")
    print(t_or_f)
    
    if t_or_f == "0":
      zeros += 1
      last_result = 0
      os.rename(files[id], files[id].rsplit("/", 1)[0]+f"/0_{zeros}.jpg")
    elif t_or_f == "1":
      ones += 1
      last_result = 1
      os.rename(files[id], files[id].rsplit("/", 1)[0]+f"/1_{ones}.jpg")
    else:
      id -= 1
      print("You have the wrong input! Please do it again!")
    print(f"id = {id+1}, zeros = {zeros}, ones = {ones}")
    id += 1
    
```
